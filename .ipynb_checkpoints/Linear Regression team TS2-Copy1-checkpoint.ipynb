### Results if we follow method 1: Taking Estimates
#### Show our calculations

```python
#Imported regression model
from sklearn.linear_model import LinearRegression

#Split predictors and response
X = train_set['Weight_Kg']
Y = train_set['avg_price_per_kg']

#Calculating x bar, y bar
x_bar = np.mean(X)
y_bar = np.mean(Y)

#Calculating Slope
b = sum((X-x_bar) * (Y-y_bar)) / sum((X-x_bar)**2)

#Calculating intercept
a = y_bar - b*x_bar

print("Slope = " + str(b))
print("Intercept = " + str(a))
```

#### Plot our results

```python
def gen_y(x_list, a, b):
    y_gen = []
    for x_i in x_list:
        y_i = a + b*x_i
        y_gen.append(y_i)

    return(y_gen)

y_gen2 = gen_y(X, a, b)

plt.scatter(X, Y)
plt.plot(X, y_gen2, color='red')
plt.show()
```
